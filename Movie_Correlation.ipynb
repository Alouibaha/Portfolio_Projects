```python
# Import required libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # Adjusts the configuration of the plots

# Read in the data

df = pd.read_csv(r'C:\Users\ALOUI\Desktop\Porfolio Files\movies.csv') # r before Location/path
```

```python
# First look at the data

df.head()
```

```python
# Looking for missing data

# Number of missing data per column
sum_missing = df.isnull().sum()
print(sum_missing)
```

```python
# Removing Missing values

df = df.dropna() 
```

```python
# Columns types

df.dtypes
```

```python
# Changing type of columns

df["budget"] = df["budget"].astype('int64')
df["gross"] = df["gross"].astype('int64')
```

```python
# Create correct year column by extarcting correct year form date of release

df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int) # Looking for patterns (4 digit(0-9))

df.head()
```

```python
# Sorting movies by gross revenue

pd.set_option('display.max_rows', None) # to fix showing all rows

df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# Remove duplicates

df = df.drop_duplicates()
```

```python
# Correlation Between All the Numeric Columns of DataFrame

df.corr()
```

```python
# Plot correlation matrix
correlation_matrix = df.corr()

sns.heatmap(correlation_matrix, annot=True)

plt.title ('Correlation Matrix For Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
# Scatter plot with budger vs gross

plt.scatter(x=df['budget'], y=df['gross'])

plt.title ('Gross Revenue Per Budget')
plt.xlabel('Gross Revenue')
plt.ylabel('Gross Budget')
plt.show()
```

```python
# Plot scatter plot Gross per Revenue using seaborn

sns.regplot(data=df, x='budget', y='gross', scatter_kws={"color": "black"}, line_kws={"color":"blue"})
```

```python
# Looking Correlation between Gross Revenue and Company

df_numerized = df

# Converting Object Type Columns to numeric (Categorical Data)
for col in df_numerized.columns:
    if (df_numerized[col].dtype == 'object'):
        df_numerized[col] = df_numerized[col].astype('category')
        df_numerized[col] = df_numerized[col].cat.codes

df_numerized.head(10)
```

```python
# Plot correlation matrix for numerized dataframe

correlation_matrix = df_numerized.corr()

sns.heatmap(correlation_matrix, annot=True)

plt.title ('Correlation Matrix For Numerized Dataframe')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
df_numerized.corr()
```

```python
# Correlation sorted for each column with others

correlationd_matrix_cat = df_numerized.corr()

corr_pairs = correlationd_matrix_cat.unstack()

sorted_corr = corr_pairs.sort_values()

sorted_corr
```

```python
# Identifing high correlation

high_corr = sorted_corr[sorted_corr > 0.5 ]

high_corr
```
